```python
tagsdict = {}
for tag in us_df['tags']:
    depiped = tag.split('|')
    for item in depiped:
        cleanitem = item.replace('"','')
        if cleanitem in tagsdict:
            tagsdict[cleanitem] += 1
        else:
            tagsdict[cleanitem] = 1
```

```python
tagsseries = pd.Series(tagsdict)
```
