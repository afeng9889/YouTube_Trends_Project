```python
import pandas as pd
import datetime
import pytz
```

```python
us_df = pd.read_csv('USvideos.csv')
us_df
gb_df = pd.read_csv('GBvideos.csv')
gb_df
```

```python
# Change trending_date to datetime format
us_df['trending_date'] = pd.to_datetime(us_df['trending_date'],format='%y.%d.%m')
```

```python
# Remove unecessary columns
del_col = ['video_id','thumbnail_link','comments_disabled','ratings_disabled','video_error_or_removed','description']
us_df = us_df.drop(del_col, axis=1)
```

```python
# Change publish_time to datetime format
us_df['publish_time'] = pd.to_datetime(us_df['publish_time'])
us_df
```

```python
df_startus=us_df['publish_time']
df_endus=us_df['trending_date']
df_startus
df_endus
```

```python
#remove timezone info from df_star
df_startus= df_startus.apply(lambda d: d.replace(tzinfo=None))
```

```python
#calculate the trending duration of each video
df_durus=df_endus-df_startus
df_durus
```

```python
#add the trending_duration column in us_df
us_df['trending_duration']=df_durus
us_df
```

```python
us_df.to_csv("USvideo.csv",index=False)
```
